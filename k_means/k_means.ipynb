```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.spatial.distance as sp_dist
import seaborn
```

```python
plt.rcParams["figure.figsize"] = (60,30)

seaborn.set(font_scale=2.5)
seaborn.set_style("whitegrid")
```

```python
X = np.loadtxt("/Users/aaronjones/Documents/github/clustering-exercises/k_means/Q4Test.txt", skiprows=1)
K = 3
```

```python
def kmeans(data, num_clusts):
    """
    Execute K-Means clustering algorithm.

    Parameters
    ----------
    data : 2d-array
        the data points to be clustered
    num_clusts : integer
        the number of clusters assumed to exist in the array

    Returns
    -------
    tuple
        contains two lists: 1) the label assignment history and 2) the centroid history
        the last elements of each list are the final results
    """
    list_labels = []
    list_centers = []

    it = 0
    while True:

        # if iteration number is zero, then random select labels
        # otherwise pull previous label assignments
        if it == 0:
            pre_labels = np.random.choice(range(num_clusts), data.shape[0])
            list_labels.append(pre_labels)
        else:
            pre_labels = list_labels[it]

        # append updated centroids to centers
        # calculate mean values within column and grouped by labels
        centers = np.array([data[pre_labels == i].mean(axis=0) for i in range(num_clusts)])
        list_centers.append(centers)

        # calculate difference between data points and centroids
        dists = sp_dist.cdist(data, centers)

        # append updated cluster assignments to labels
        # calculate minimum distance within row
        labels = np.argmin(dists, axis=1)
        list_labels.append(labels)

        # if labels remain the same, then calculate centers and break
        # otherwise increase iteration number
        if (pre_labels == labels).all():
            centers = np.array([data[labels == i].mean(axis=0) for i in range(num_clusts)])
            list_centers.append(centers)
            break
        else:
            it += 1

    return (list_labels, list_centers)
```

```python
labels, centers = kmeans(data=X, num_clusts=K)
```

```python
M = len(labels)
print("Algorithm converged in {} steps.".format(M))
```

```python
# plotting convergence path, including final results
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

for idx, i in enumerate([0, (M - 1) // 2, M - 1]):
    for j in range(K):
        X_filt = X[labels[i] == j]
        axes[idx].set_title("Iteration: " + str(i))
        axes[idx].scatter(X_filt[:,0], X_filt[:,1])
        axes[idx].scatter(centers[i][:,0], centers[i][:,1], c="black", s=100)
```
